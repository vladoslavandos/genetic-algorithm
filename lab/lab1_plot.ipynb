```python
from matplotlib import pyplot as plt
from matplotlib.ticker import NullFormatter, FixedLocator
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np
```

## График зависимости времени нахождения одного решения методом полного перебора от amax

Логарифмическая шкала по amax

```python
amax = [144715, 360974, 1048576, 3697661, 16777216, 106528681, 1073741824]
mse = [0.428460, 0.525017, 0.608460, 0.782811, 0.898886, 0.902000, 0.842872]
time = [0.183578, 0.275643, 0.370224, 0.612793, 0.807996, 0.813604, 0.710434]

plt.subplots(figsize=(10,6))
plt.errorbar(amax, time, mse)
#plt.plot(amax, time)
plt.xlabel('log(amax)')
plt.xscale('log')
plt.ylabel('time of work')
plt.grid()
```

D(a_max) = n / log2(amax)


```python
def forward(x):
    return 24 / np.log2(x)

def inverse(x):
    return 2 ** (24 / x)

fig, axs = plt.subplots(figsize=(10,6))

#print(forward(amax))

plt.errorbar(forward(amax), time, mse)
#axs.set_xscale('function', functions=(forward, inverse))
#plt.plot(amax, time)
plt.xlabel('D(amax) = n / log2(amax)')
plt.ylabel('time of work')
plt.grid()
```

С интерполяцией

```python
new_x = np.linspace(min(amax), max(amax), 1000)
a_BSpline = make_interp_spline(amax, time)
new_y = a_BSpline(new_x)

plt.subplots(figsize=(10,6))
plt.plot(new_x, new_y)
plt.xlabel('amax')
plt.xscale('log')
plt.ylabel('time of work')
plt.grid()
```
